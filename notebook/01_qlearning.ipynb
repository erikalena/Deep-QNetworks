## Q-Learning in Snake


Start by standard configuration we saw for the simple example of the cliff presented in Sutton and Burto Ex.6.6.
This time we will use a different environment, the snake game.

```python
import sys
sys.path.append('../src_code/qnetworks/')
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from qlearning import *
```

```python
# size of the gridworld
Lx = 10
Ly = 10

env = GridWorldEnv(Lx,Ly)
env.render()
```

```python
# learn policies
n_steps = env.learn_policy()
```

```python
n_steps = np.array(n_steps)
# plot convergence
plt.figure()
plt.plot(np.arange(len(n_steps)), n_steps, label='V')
plt.xlabel('Episodes', fontsize=16)
plt.ylabel('Number of steps', fontsize=16)
plt.legend()
plt.show()
```
